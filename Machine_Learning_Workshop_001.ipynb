# What is Machine Learning?

Arthur Samuel.jpeg

**"A field of study that gives computers the ability
to learn without being explicitly programmed." - Arthur Samuel (1959)**

Broadly, ML algorithms fall into the following categories:

    1) Supervised Learning
    2) Unsupervised Learning
    3) Other...

# 1) Supervised Learning

    Def: Algorithms that learn via mappings from inputs (X) to outputs (Y), or
         output labels. Or, in other words, algorithms that learn from training
         data sets (i.e. an answer sheet that contains the "correct" answers).

Real World Examples:

    Spam filters, speech recognition, machine translation,
    online ads, self-driving cars, visual inspection, etc.

Common Types of Supervised Learning Models:

    Linear Regression, Random Forest, Decision Trees, Support
    Vector Machines (SVM's), and K-Nearest Neighbors (K-NN).

# **Broad Subcategories of Supervised Learning:** 

**Regression Algorithm:**

    A type of supervised learning algorithm that learns to
    predict numbers out of infinitely many possible numbers.

Example: "Predicting house price using regression analysis based on square footage and other house characteristics."

**Classification Algorithm:**

    A type of supervised learning algorithm that learns to predict out 
    of a small number of possible outputs, categories, or classes.

Example: "Detecting breast cancer using data to correctly classify tumors as either malignant or benign based on tumor size."

# 2) Unsupervised Learning

    Def: Algorithms used to detect structures, patterns, labels, or other
         interesting features of data without being provided "correct" answers.
         (i.e., data includes only inputs (X) and no output labels (Y)).

Real World Examples:

    Anomaly detection, customer behavior prediction,
    recommendation engines, noise removal in data sets, etc.

# Broad Subcategories of Unsupervised Learning:

    Clustering, Association, Density Estimation, Dimensionality
    Reduction, and Singular Value Decomposition (SVD). 

Clustering Algorithm: 

    Def: A clustering algorithm is an unsupervised algorithm where inputs are 
         not labeled and problem solving is based on the experience that
         the algorithm gains out of solving similar problems as a training
         schedule. The goal is to group similar data points together.

Specific Algorithms:

    K-Means Clustering, Principal Component Analysis (PCA), and Hierarchical clustering.

# Practical Examples of Unsupervised Learning:

1.   Every day, Google News attempts to create lists of similarly related articles (i.e., clusters).

2.   Using DNA microarray data to uncover assocations between genes and an individual's behavior.

3. Creating segments of customers based on customer characteristics to make better recommendations.

# **Linear Regression Model - Univariate Case**

Given a training data set: $x^{(i)},y^{(i)}$ where there appears to be preliminary evidence of a linear relationship between random variables x and y, we can use linear regression to model the data.

For example, suppose we create a scatterplot of the data, and it looks like this:

Linear_Scatterplot.png

The equation for a linear regression with one feature is:
$$ $$
$$\hat{y} = \beta_0 + \beta_1 \cdot x \tag{Eq. 1.1}$$

$$ $$
- $\hat{y}$ : predicted values of y
- x : the feature or input
- $\beta_0$ : the intercept parameter
- $\beta_1$ : the slope parameter

The actual values of y (`i` = specific example) can be calculated using the following equation:
$$ $$
$$y^{(i)} = \beta_0 + \beta_1 \cdot x^{(i)} + \epsilon^{(i)} \tag{Eq. 1.2}$$
$$ $$

where epsilon is the **"error term"** measuring the distance between $\hat{y}$ and y for a specific observation $x^{(i)},y^{(i)}$ given by the following equation:
$$ $$
$$\epsilon^{(i)} = \hat{y}^{(i)} - y^{(i)} \tag{Eq. 1.3}$$

**$\hat{y}$** is the **predicted value of y** (i.e., the response variable) in a linear regression model. To distinguish between values of **$\hat{y}$** and values of **y**, we can refer to the former as **"predicted values"** and the latter as **"actual values"** or **"real values"**.

# Model Fitting

The goal is to select values for the parameters $\beta_0$ and $\beta_1$ that optimally **"fit"** the data. Using linear regression, we can create a **"line of best fit"** that passes most closely through the data points. Seeing a visual representation may be the easiest way to begin to understand this concept:


Model_Fitting.png

In an effort to "optimize" the parameters of our linear regression model (i.e., obtain better predictions) we can quantify the average amount by which each prediction "misses its mark" and then **minimize** that quantity. To do this, we will introduce the concept of a **"cost function"**. The first cost function introduced in this series utilizes **mean squared error**.

# Mean Squared Error (MSE)

To begin, the first step is to square the errors (i.e., **"residuals"**) - a technique from traditional statistics used to obtain non-negative weights - enabling us to obtain a quantity that measures the accuracy of our model. In machine learning, this quantity is called the **"cost"** or **"loss"**.


$$ $$
$$(\hat{y}^{(i)} - y^{(i)})^2 \tag{a}$$
$$ $$
- Recall that $\hat{y}^{(i)} = f_{\beta_0,\beta_1}(x^{(i)}) = \beta_0 + \beta_1 \cdot x^{(i)}$

The second step is to obtain the sum squared errors (SSE):
$$ $$
$$\sum\limits_{i = 1}^{m} (\hat{y}^{(i)} - y^{(i)})^2 \tag{b}$$
$$ $$

The third step is to divide the sum of squared errors (SSE) by the number of training examples (m). The resulting function is interpreted as "the average cost" of the model and is referred to as the **mean squared error**:
$$ $$
$$ \frac{1}{m} \sum\limits_{i = 1}^{m} (\hat{y}^{(i)} - y^{(i)})^2 \tag{c}$$

The final step is to divide the mean squared error (MSE) by 2:
$$ $$
$$ \frac{1}{2m} \sum\limits_{i = 1}^{m} (\hat{y}^{(i)} - y^{(i)})^2 \tag{d}$$
$$ $$
- Dividing by 2 makes the differentiation of the cost function easier during gradient descent.

# Cost Function - Mean Squared Error (MSE)

From part (d), we obtain the cost function, $J(\beta_j)$, for linear regression of one variable:
$$ $$
  $$J(\beta_0,\beta_1) = \frac{1}{2m} \sum\limits_{i = 1}^{m} (f_{\beta_0,\beta_1}(x^{(i)}) - y^{(i)})^2 \tag{Eq. 1.4}$$ 
 
where:
$$ $$
  $$f_{\beta_0,\beta_1}(x^{(i)}) = \beta_0 + \beta_1 \cdot x^{(i)} \tag{Eq. 1.5}$$
$$ $$
- $f_{\beta_0,\beta_1}(x^{(i)})$ is our prediction for example $i$ using parameters $\beta_0,\beta_1$.  
- $(f_{\beta_0,\beta_1}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the prediction and the actual value (or **target**).   
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(\beta_0,\beta_1)$.  


# Gradient Descent (General Case)

Since we want to build a model that makes accurate predictions, i.e., our goal is to build a model with parameters $\beta_0,\beta_1$ that minimizes $J(\beta_j)$. Hence:
$$ $$
$$ \min_{\beta_0,\beta_1} J(\beta_0,\beta_1) $$
$$ $$
- At this point, we are using just two parameters, but we will eventually extend this model out to `k` parameters

Because the MSE cost function is convex (i.e., it has an absolute minimum), we can use partial derivatives of $J(\beta)$ with respect to each parameter to estimate the optimal parameters $\beta_0,\beta_1$. Gradient descent is a systematic algorithm for iteratively solving for the model parameters. In the general case, the **gradient descent** algorithm follows the following form:
$$ $$
$ Repeat $ $ \{ $
\
$ \beta_j := \beta_j - \alpha \cdot \frac{\partial}{\partial \beta_j} J(\beta) \tag{Eq. 1.6}
\ $
\
$ \} $
$$ $$
- `j` : a given parameter (e.g., 0, 1, 2, ...)
- $\alpha$ : the learning rate (this is an example of a hyperparameter)
- $\frac{\partial}{\partial \beta_j}J(\beta)$ : the partial derivative of the cost function, $J(\beta)$, with respect to parameter $\beta_j$
$$ $$


# Gradient Descent (Univariate Case)

This gradient descent algorithm updates each parameter synchronously at each and every iteration as opposed to updating one parameter before updating another parameter in sequential order.
$$ $$
In the univariate case, the intercept and slope paramters are updated in the following fashion:
$$ $$
$ Repeat $ $ \{ $
$$ $$
$$ \beta_0 := \beta_0 - \alpha \cdot \frac{\partial}{\partial \beta_0} J(\beta_0,\beta_1)
\ $$
$$ \tag{Eq. 1.7} $$
$$ \beta_1 := \beta_1 - \alpha \cdot \frac{\partial}{\partial \beta_1} J(\beta_0,\beta_1) $$
$$ $$
$ \} $


# Differentiating the Cost Function

Let's take the derivative of $J(\beta)$ w.r.t. $\beta_j$ :
$$ $$
$$
\begin{align}
\frac{\partial}{\partial \beta_j} J(\beta) 
& = \frac{\partial}{\partial \beta_j} \frac{1}{2m} \sum\limits_{i = 1}^{m} (f_{\beta}(x^{(i)}) - y^{(i)})^2\\
\\
& = 2 \cdot \frac{1}{2m} \sum\limits_{i = 1}^{m} (f_{\beta}(x^{(i)}) - y^{(i)}) \cdot \frac{\partial}{\partial \beta_j} (f_{\beta}(x^{(i)}) - y^{(i)})\\
\\
& = \frac{1}{m} \sum\limits_{i = 1}^{m} (f_{\beta}(x^{(i)}) - y^{(i)}) \cdot \frac{\partial}{\partial \beta_j} \beta_j x_j\\
\\
& = \frac{1}{m} \sum\limits_{i = 1}^{m} (f_{\beta}(x^{(i)}) - y^{(i)}) \cdot  x_j^{(i)} \tag{Eq. 1.8}\\
\end{align}
$$
$$  $$

Therefore, the partial derivatives used in gradient descent for the univariate case are:
$$ $$
$$ \frac{\partial}{\partial \beta_0} J(\beta_0,\beta_1) =  \frac{1}{m} \sum\limits_{i = 1}^{m} (f_{\beta}(x^{(i)}) - y^{(i)}) \tag{Eq. 1.9}$$
$$ $$
$$ \frac{\partial}{\partial \beta_1} J(\beta_0,\beta_1) =  \frac{1}{m} \sum\limits_{i = 1}^{m} (f_{\beta}(x^{(i)}) - y^{(i)}) \cdot x^{(i)} \tag{Eq. 1.10}$$
$$ $$
- Eq. 1.9 is the partial derivative of $J(\beta)$ w.r.t $\beta_0$ (i.e., the intercept parameter)
- Eq. 1.10 is the partial derivative of $J(\beta)$ w.r.t $\beta_1$ (i.e., the slope parameter)

Now let's create a function that generates predictions using the linear model: $f_{\beta}(x^{(i)}) = \beta_0 + \beta_1 \cdot x^{(i)}$

```python
# Linear Model
def linear_model_predict(x, b0, b1):
    """
    Computes the prediction of a linear model
    
    Args:
      x (ndarray (m,))   : Data, m examples 
      b0,b1 (scalars)    : model parameters  
    
    Returns
      y (ndarray (m,))   : target values
    """
    m = x.shape[0]
    f_b = np.zeros(m)
    for i in range(m):
        f_b[i] = b0 + b1 * x[i]
    return f_b
```

Clearly, our predictions could be better... Let's calculate the cost of our current model (first manually and then with a function).

```python
# Calculate the model errors (i.e., residuals)
error = y_pred - y_train
print(error) # Negative values indicates the model under-predicts

# Calculate the squared errors
sq_error = error**2
print(sq_error)

# Calculate the sum of squared error
SSE = sum(sq_error)
print(SSE)

# Calculate the mean squared error
MSE = (1/m) * SSE
print(MSE)

# Calculate the total cost of the model
total_cost = (1/2) * MSE
print(total_cost)
```

Now, let's create a function to compute the cost:

```python
# Create function to compute costs of linear model
def compute_cost(x, y, b0, b1):
    """
    Computes the cost function for univariate linear regression.
    
    Args:
      x (ndarray (m,)) : Data, m examples 
      y (ndarray (m,)) : target values
      b0,b1 (scalars)  : model parameters  
    
    Returns
        total_cost (float) : The cost of using b0,b1 as the parameters for
                             linear regression to fit the data points in x and y
    """
    m = x.shape[0] 
    cost = 0

    for i in range(m):
        f_b = b0 + (b1 * x[i])
        cost = cost + (f_b - y[i])**2
    total_cost = 1 / (2 * m) * cost

    return total_cost
```

Now let's create a few functions so that we can run gradient descent. The first function calculates the gradients using partial derivatives.

```python
# Create function to compute gradients
def compute_gradient(x, y, b0, b1): 
    """
    Computes the gradient for linear regression 

    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      b0,b1 (scalars) : model parameters

    Returns
      dj_db0 (scalar): The gradient of the cost w.r.t. the parameter b0
      dj_db1 (scalar): The gradient of the cost w.r.t. the parameter b1  
     """
    # Number of training examples
    m = x.shape[0]    
    dj_db0 = 0
    dj_db1 = 0

    for i in range(m):  
        f_b = b0 + (b1 * x[i])
        dj_db0_i = f_b - y[i] 
        dj_db1_i = (f_b - y[i]) * x[i] 
        dj_db0 += dj_db0_i
        dj_db1 += dj_db1_i 
    dj_db0 = (1/m) * dj_db0
    dj_db1 = (1/m) * dj_db1
    
    return dj_db0, dj_db1
```

**Function: "compute_gradient"**
$$ $$
This function computes the gradients of the cost function for linear regression w.r.t. the model parameters **b0** (i.e., $\beta_0$) and **b1** (i.e., $\beta_1$). The input variables are:
$$ $$

*   x : an ndarray of shape (m,) representing the data with m examples.
*   y : an ndarray of shape (m,) representing the target values.
*   b0, b1 : scalars representing the model parameters $\beta_0,\beta_1$. $$ $$

The function first initializes **dj_db0** (i.e., $\frac{\partial J}{\partial \beta_0}$) and **dj_db1** (i.e., $\frac{\partial J}{\partial \beta_1}$) to zero, which will later accumulate the gradient values. It then loops through each training example in the input **x** and computes the predicted output using the current model parameters **b0** and **b1**. The gradients of the cost function with respect to **b0** and **b1** are then computed for each training example using the predicted output and the corresponding target value.
$$ $$
Finally, the accumulated values of **dj_db0** and **dj_db1** are divided by the number of training examples to obtain the average gradient values, which are then returned as a tuple of two scalars (**dj_db0**, **dj_db1**).

Now, finally we need to create a function that computes gradient descent utilizing both the **"compute_cost"** and **"compute_gradient"** functions:

```python
# Create gradient descent function
def gradient_descent(x, y, b0_in, b1_in, alpha, num_iters, cost_function, gradient_function, print_progress=True): 
    """
    Performs gradient descent to fit b0,b1. Updates b0,b1 by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      x (ndarray (m,))       : Data, m examples 
      y (ndarray (m,))       : target values
      b0_in, b1_in (scalars) : initial values of model parameters  
      alpha (float)          : learning rate
      num_iters (int)        : number of iterations to run gradient descent
      cost_function          : function to call to produce cost
      gradient_function      : function to call to produce gradient
      
    Returns:
      b0 (scalar)      : Updated value of intercept parameter after running gradient descent
      b1 (scalar)      : Updated value of slope parameter after running gradient descent
      J_history (List) : History of cost values
      p_history (list) : History of parameters [b0,b1] 
      """
    # An array to store cost J and parameters at each iteration primarily for graphing later
    J_history = []
    p_history = []
    b0 = b0_in
    b1 = b1_in

    for i in range(num_iters):
        # Calculate the gradient and update the parameters using gradient_function
        dj_db0, dj_db1 = gradient_function(x, y, b0, b1)   

        # Update Parameters using equation (3) above
        b0 = b0 - alpha * dj_db0                            
        b1 = b1 - alpha * dj_db1                  

        # Save cost J at each iteration
        if i<100000: # prevent resource exhaustion 
            J_history.append(cost_function(x, y, b0, b1))
            p_history.append([b0,b1])

        # Print cost every at intervals 10 times or as many iterations if < 10
        if print_progress and (i% math.ceil(num_iters/10) == 0):
            print(f"Iteration {i:5}:  |  Cost {J_history[-1]:15.1f} ",
                  f"|  dj_db0: {dj_db0:9.1e}, dj_db1: {dj_db1:9.1e} ",
                  f"|  b0:{b0:9.2f}, b1:{b1:9.2f}")
 
    return b0, b1, J_history, p_history
```

**Function: "gradient_descent"**
$$ $$
The function takes as input the following arguments:
$$ $$
* x : an ndarray of shape (m,) representing the input data (predictor variable).
* y : an ndarray of shape (m,) representing the target values (response variable).
* b0_in, b1_in : initial values of the intercept and slope parameters of the linear regression model.
* alpha : the learning rate used in the gradient descent algorithm (hyperparameter).
* num_iters : the number of iterations to perform gradient descent (hyperparameter).
* cost_function : a pre-defined function to compute the cost of the linear regression model.
* gradient_function : a pre-defined function to compute the gradients of the cost function with respect to the model parameters.
* print_progress : a boolean flag indicating whether to print the progress of the gradient descent algorithm. $$ $$ 

The function performs the following steps:
$$ $$
* Initializes the intercept and slope parameters **b0** and **b1** to **b0_in** and **b1_in**, respectively.
* For each iteration *i* in **num_iters**, computes the gradients of the cost function w.r.t. the model parameters using the gradient_function.
* Updates the model parameters **b0** and **b1** by subtracting alpha times the gradients.
* Computes the cost of the linear regression model using the "cost_function" and appends it to the **J_history** list.
* Appends the current parameter values **[b0, b1]** to the **p_history** list.
* If **print_progress** is True, prints the current cost, gradients, and parameter values every 10% of num_iters.
* Returns the updated parameters **b0** and **b1**, the history of cost values **J_history**, and the history of parameter values **p_history**. $$ $$

Now we are ready to run our first machine learning model using gradient descent!

```python
"""
Gradient descent setup & execution
"""

# Initialize parameters
b0_in = 0
b1_in = 0

# Adjust some gradient descent settings
iterations = 10000
alpha = 0.0125

# Run gradient descent to determine optimal values for b0 and b1
b0_final, b1_final, J_hist, p_hist = gradient_descent(x_train, y_train, b0_in, b1_in, alpha, 
                                                    iterations, compute_cost, compute_gradient,
                                                    print_progress=True)
```
